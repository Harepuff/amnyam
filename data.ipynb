cianparser загружает данные сразу в переменную, но для наглядности вся информация была записана в csv для дальнейшей работы с ней

```python
df = pd.read_csv('cian.csv', sep=';')
df.head()
```

```python
df.drop(['author', 'url', 'location', 'deal_type', 'accommodation_type', 'house_number', 'commissions'], axis=1, inplace=True)
        # данные, выбранные для удаления - ни на что не влияют;
        # имён авторов заявок и номеров домов слишком много разных для какого-либо вменяемого анализа,
        # url у всех разный и не нужен,
        # а локация, комиссия (везде 0) тип недвижимости и тип сделки у всех одинаковые
df.drop(df[df.rooms_count == -1].index, inplace=True)
        # удалено несколько строк, где не понятно, сколько квартир в доме
        # было решено не заменять на "1" или какую-либо медианную, т.к.
        # при ручном рассмотрении, под "-1" могут скрываться как апартаменты, так и квартиры,
        # где слишком много комнат (свыше 5)
df.head()
```

```python
df.to_csv('data_optimized.csv', index=False)
```
